```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime
from datetime import date,timedelta
```

```python
apple = pd.read_csv('apple.csv', index_col='Date', parse_dates=True)
msft = pd.read_csv('msft.csv', index_col='Date', parse_dates=True)
```

```python
apple.head()
```

```python
apple['Daily Returns'] = apple['Adj Close'].pct_change(1)
msft ['Daily Returns'] = msft['Adj Close'].pct_change(1)
```

```python
apple = apple.dropna()
msft = msft.dropna()
```

```python
def compute_sortino_ratio (data, threshold = 0, risk_free_rate = 0):
    mean_data = data['Daily Returns'].mean()
    #Always lower values than treshold
    downside = data[data['Daily Returns']<threshold]['Daily Returns']
    std = downside.std()
    sortino_ratio = (mean_data - risk_free_rate)/std

    return sortino_ratio * (252**0.5)
```

```python
sortino_apple = compute_sortino_ratio(apple)
sortino_apple
```

```python
sortino_msft = compute_sortino_ratio(msft)
sortino_msft
```
